```python
England_Tables
```

```python
#Remove non relevant columns
England_Tables.drop(columns = ['Top Team Scorer'],inplace = True)
```

```python
#Rename of columns
England_Tables.rename(columns = {'Rk':'Position','MP':'Played Games','W':'Wins','D':'Draws','L':'Losed','GF':'Goals For',
                                 'GA':'Goals Against','GD':'Goals Difference','Pts':'Points'},inplace= True)
```

```python
#Change data type of specific columns for aggregation 
England_Tables['Scored Goals'] = England_Tables['Scored Goals'].astype('int64')
England_Tables['Attendance'] = England_Tables['Attendance'].astype('int64')*1000
```

```python
#Adding aggregation column with result round
England_Tables['% of Teams Score'] = ((England_Tables['Scored Goals']/England_Tables['Goals For'])*100).round(1)
```

```python
England_Tables
England_Tables.to_excel('C:\PROJECT-2\BI Files\England_Tables.xlsx')
```

```python
#ENGLAND - Transfers CONCAT
#Import data from xlsx file
file3 = pd.read_excel('England-Transfer Historie.xlsx',usecols = 'c:g,i:k',sheet_name=None)
```

```python
#Creating DataFrame
England_Transfers= pd.concat(file3)
```

```python
#Changing format for new columns
England_Transfers['Spent money(mln € )'] = England_Transfers['Spent money(mln € )'].astype('float')
England_Transfers['Income money(mln € )'] = England_Transfers['Income money(mln € )'].astype('float')
```

```python
#Adding aggregation column
England_Transfers['Balance'] = England_Transfers['Spent money(mln € )'] - England_Transfers['Income money(mln € )']
```

```python
#Creating EXCEL file in a destination folder
England_Transfers
England_Transfers.to_excel('C:\PROJECT-2\BI Files\England_Transfers.xlsx')
```
